```python
from IPython.display  import clear_output
```

```python
def agregar_tarifa(tarifa,mins):
  clear_output()
  tarifa.append(mins)
  print("El numero de horas {} ha sido agregado a la tarifa ".format(mins))
  return tarifa

def mostrar_tarifa(tarifa):   
  clear_output()                  
  if len(tarifa)>0:
    for mins in tarifa:
      print(" - {}".format(mins))
  else:
    print("No hay tarifa relacionada")

def eliminar_tarifa(carrito,prod):
  clear_output()
  try:            
    carrito.remove(prod)
    print(" {} ha sido removido".format(prod))
  except:
    print("")
    
def vaciar_tarifa(carrito):
  clear_output()
  carrito.clear()
  print("Se han removido todos los productos. ")

def pagar_tarifa(carrito):
  total=0
  precio = 0.00
  for prod in carrito:
    #precio=float(input("¿Cual es el precio de {}?".format (prod)))
    total+=int(prod) #precio #total=total+precio
  
  #tomar los primeros 60 mins

  #total = total - 60
  total = total - 1
  precio = precio + 25.00

  #calcular horas
  #agregar_extra = total % 60 != 0

  #horas = total // 60

  agregar_extra2 = total > 8

  precio = precio + (total * 15.00)

  #precio += 15.00 if agregar_extra else 0
  precio += 200.00 if agregar_extra2 else 0

  print("Total: {}".format(precio))
```

```python
def principal():  
  terminado=False
  tarifa=[]
  while not terminado :
    res=input("¿Que te gustaria hacer?   agregar / eliminar / mostrar / vaciar / salir: ").lower()
    if res=="salir":
      if len(tarifa)>0:
        pagar_tarifa(tarifa)
      else:
        clear_output()
        print("Gracias por usar nuestro servicio.")
      terminado=True
      
    elif res=="agregar":
      mins=input("¿Cuantas horas vas a agregar?")
      tarifa=agregar_tarifa(tarifa,mins)

    elif res=="eliminar":
      mostrar_tarifa(tarifa)
      mins=input("¿Que prodcuto quieres eliminar?")
      tarifa=eliminar_tarifa(tarifa,mins)
    elif res=="mostrar":
      mostrar_tarifa(tarifa)
    elif res=="vaciar":
      carrito=vaciar_tarifa(tarifa)
    else:
      print("Comando no valido, intente nuevamente")  
      
```
